#3. Handling Missing Data
Identify if there are any missing values in the dataset. Replace missing values in the
horsepower column with the column's median.

```python
import pandas as pd
df =pd.read_csv('/content/auto-mpg.csv')
# Check missing values
print(df.isna().sum())
```

```python
# Convert to numeric (handle '?')
df['horsepower']=pd.to_numeric(df['horsepower'],errors='coerce')
```

```python
# Check again the missing values
print(df.isna().sum())
```

```python
# Find Median
median_hp=df['horsepower'].median()
print(median_hp)
```

```python
# Fill missing values with Median
df['horsepower']=df['horsepower'].fillna(median_hp)
```

```python
# Verify
print(df.isna().sum())
```

#10. Finding Anomalies
Identify potential outliers in the mpg column using the Interquartile Range (IQR)
method. Specifically:
● Calculate the IQR for mpg.
● Define outliers as values less than Q1 - 1.5 * IQR or greater than Q3 +
1.5 * IQR.
● Create a DataFrame of cars classified as outliers, displaying car_name, mpg,
# and model_year.

```python
Q1 = df['mpg'].quantile(0.25)
Q3 = df['mpg'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = df[(df['mpg'] < lower_bound) | (df['mpg'] > upper_bound)]
```

```python
outliers_df = outliers[['car name', 'mpg', 'model year']]
print(outliers_df)
```

#5. Broadcasting
Increase all horsepower values by 10% and store the updated values in a new
NumPy array. Handle missing data (if any) by replacing it with the mean of the
column before applying the increase.

```python
# Find Column Indexes
df.columns
```

```python
# Extract Horsepower Column
hp_array = df['horsepower'].to_numpy()
```

```python
# Handle missing values
# Convert hp_array to numeric first, coercing non-numeric values to NaN
hp_array = pd.to_numeric(hp_array, errors='coerce')
# Now np.isnan can be applied
np.isnan(hp_array).sum()
np.int64(0)
```

```python
# Replace missing values with Mean
mean_hp=np.nanmean(hp_array)
hp_array=np.where(np.isnan(hp_array),mean_hp,hp_array)
```

```python
# Apply 10% Increase (Broadcasting)
updated_hp=hp_array*1.10
```

```python
print(updated_hp[:10])
```

#10. Conditional Aggregates
Calculate the mean mpg for cars grouped by the number of cylinders using NumPy
techniques.

```python
# Find Column Indexes
df.columns
```

```python
# Extract Columns
mpg = df_np[:, 0].astype(float)
cylinders = df_np[:, 1].astype(int)
```

```python
# Get Unique Cylinder Values
unique_cyl = np.unique(cylinders)
```

```python
# Loop and Compute Mean
for cyl in unique_cyl:
    mask = cylinders == cyl
    avg_mpg = np.mean(mpg[mask])
    print(f"Cylinders: {cyl}, Avg MPG: {avg_mpg}")
```
