#2. Weather Data Analyzer
a. Analyze weather data (temperature, humidity, etc.) for a given region,
b. Use variables and data types to store weather data,
c. Implement conditional statements to determine weather conditions
(sunny, rainy, etc.),
d. Use functions to calculate average temperature and humidity, and print
results


```python
# Weather Data Analyzer

# Sample data
temperatures = [30, 32, 31, 29, 28]
humidity = [60, 65, 70, 75, 80]

# Function to calculate average
def calculate_average(data):
    return sum(data) / len(data)

# Function to determine weather condition
def weather_condition(avg_temp, avg_humidity):
    if avg_temp > 30 and avg_humidity < 70:
        return "Sunny"
    elif avg_humidity >= 70:
        return "Rainy"
    else:
        return "Cloudy"

# Calculate averages
avg_temp = calculate_average(temperatures)
avg_hum = calculate_average(humidity)

# Determine condition
condition = weather_condition(avg_temp, avg_hum)

# Output
print("Average Temperature:", avg_temp)
print("Average Humidity:", avg_hum)
print("Weather Condition:", condition)
```
