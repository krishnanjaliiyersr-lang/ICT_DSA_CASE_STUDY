#Data Acquisitionand Analysis of SpaceX Launch Data

###Load SpaceX Launch Data

Step 1: Load SpaceX Launch Data from API
● Use https://api.spacexdata.com/v4/launches
● Extract relevant columns: name, date_utc, success, details, rocket
● Convert date_utc to datetime and extract the year

```python
import pandas as pd
import requests

# URL for SpaceX launches API
url_launches = "https://api.spacexdata.com/v4/launches"

# Fetch data from the API
response = requests.get(url_launches)

# Convert API response to JSON format (Python dictionary/list)
launch_data = response.json()

# Convert JSON data into a Pandas DataFrame for better readability
launch_df = pd.DataFrame(launch_data)

# Keep only the required columns
# Double brackets are used to select multiple columns
launch_df = launch_df[['name', 'date_utc', 'success', 'details', 'rocket']]

# Convert 'date_utc' from text (string) to datetime format
# This helps in performing date-based operations
launch_df['date_utc'] = pd.to_datetime(launch_df['date_utc'])

# Extract only the year from the datetime column
# .dt is used to access datetime properties
launch_df['year'] = launch_df['date_utc'].dt.year

# Display the first 5 rows of the DataFrame
launch_df.head()
```

### Load Rocket Metadata
● Use https://api.spacexdata.com/v4/rockets
● Extract id, name, type, active, and stages

```python
url_rockets= "https://api.spacexdata.com/v4/rockets"
response= requests.get(url_rockets)
rocket_data= response.json()
rocket_df= pd.DataFrame(rocket_data)

# Extract column
rocket_df= rocket_df[['id','name','type','active','stages']]

rocket_df.head()
```

###Merge Launch & Rocket Data

###Join the two DataFrames on rocket ID using

```python
#left_on → column name in the left DataFrame
#right_on → column name in the right DataFrame
#how ='left'It means LEFT JOIN
#Keep ALL rows from the left table (launch_df)
#Add matching data from the right table

merged_df=pd.merge(launch_df,rocket_df,left_on='rocket',right_on='id',how='left')
merged_df.head()
```

#Add Simulated Country Information
● Add a new column country and randomly assign one of these values:
['USA', 'Russia', 'India', 'China', 'France']

```python
import random

# this creates a new column country,
#random.choice(countries):(Picks one random country from the list),

countries=['USA','Russia','India','China','France']
merged_df['country']=[random.choice(countries) for _ in range (len(merged_df))]
merged_df.head(5) # show the top 5 row
```

### Store Merged Data in SQLite3
● Use sqlite3 to create a connection and save the merged DataFrame as a table named
launches
● Table should contain all merged columns including country

```python
import sqlite3

conn = sqlite3.connect("spacex.db")

merged_df.to_sql("launches", conn, if_exists="replace", index=False)

print("Data stored successfully!")
```

```python
#To show table contain all merged values
query = "SELECT * FROM launches LIMIT 5;"
pd.read_sql(query, conn)
```

# Run SQL Queries


###Launches by Country

```python
q1 = """
SELECT country, COUNT(*) AS total_launches
FROM launches
GROUP BY country
ORDER BY total_launches DESC;
"""

pd.read_sql(q1, conn)
```

### Year with Highest Launches

```python
q2 = """
SELECT year, COUNT(*) AS total_launches
FROM launches
GROUP BY year
ORDER BY total_launches DESC
LIMIT 1;
"""

pd.read_sql(q2, conn)
```

### Top 5 Missions by Launch Count

```python
q3 = """
SELECT name_x, COUNT(*) AS launch_count
FROM launches
GROUP BY name_x
ORDER BY launch_count DESC
LIMIT 5;
"""

pd.read_sql(q3, conn)
```
